### Resolucion Ejercicio 6

```python
vdf.columns
```

```python
ejer_6x = vdf.sort_values(['category','item','variety'], ascending=False)
```

```python
ejer_6x['rank'] = ejer_6x.groupby(['category','item','variety'])['year'].rank(method='first', ascending=False)
```

```python
ejer_6x = ejer_6x.loc[ejer_6x["rank"] == 1.0,:].reset_index(drop=True)
```

```python
ejer_6x
```

```python
ejer_6y = vdf.sort_values(['category','item','variety'], ascending=True)
```

```python
ejer_6y['rank'] = ejer_6y.groupby(['category','item','variety'])['year'].rank(method='first', ascending=True)
```

```python
ejer_6y = ejer_6y.loc[ ejer_6y['rank'] == 1.0,:].sort_values(['category','item','variety'], ascending=False).reset_index(drop=True)
```

```python
ejer_6y
```

```python
ejer_6x = ejer_6x.drop(['date','unit','year','year-month','rank'], axis=1)
```

```python
ejer_6x
```

```python
ejer_6y = ejer_6y.drop(['date','unit','year','year-month','rank'], axis=1)
```

```python
ejer_6y
```

```python
ejer_6= ejer_6x
```

```python
ejer_6['percentage-change-between-first-and-last-price'] = (ejer_6x['price']- ejer_6y['price'])*100
```

```python
ejer_6 = ejer_6.drop(['price'], axis=1)
```
